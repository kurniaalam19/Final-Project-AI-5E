```python
# CROP IMAGE
crop = img[100:200, 100:500]

cv2_imshow(crop)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(crop.shape)
```

```python
# COPY IMAGE
copy = img.copy()
print(copy.shape)
```

```python
# RESIZE IMAGE(IGNORE ASPECT RATIO)
resized = cv2.resize(img, (400,400))

cv2_imshow(resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# RESIZE IMAGE(MEMPERTAHANKAN ASPECT RATIO)
r = 400/img.shape[1]
dim = (400,int(img.shape[0]*r))
resized = cv2.resize(img,dim)

cv2_imshow(resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# ROTATING IMAGE
(h,w) = img.shape[:2]
center = (w/2,h/2)

M = cv2.getRotationMatrix2D(center, 90, 1.0)
rotated = cv2.warpAffine(img, M, (w,h))

cv2_imshow(rotated)
```

```python
# ADJUST IMAGE CONTRAST
import numpy as np
adjusted = cv2.addWeighted(resized, 1.5, np.zeros(resized.shape, resized.dtype), 0, -100)

cv2_imshow(resized)
cv2_imshow(adjusted)
```

```python
# DETECT EDGES
edgess = cv2.Canny(resized, 100,200)

cv2_imshow(resized)
cv2_imshow(edgess)
```
